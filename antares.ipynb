```python
import pandas as pd
```

```python
df = pd.read_json('data_antares.json')
df
```

```python
df.set_index('time', inplace=True)
df.drop(columns=['dataUri'])
```

```python
df[['Roll Rate', "Pitch Rate", "Yaw Rate"]] = df['content'].str.split(',', expand=True)
```

```python
df = df.drop(columns=['dataUri', 'content'])
```

```python
df
```

```python
for col in ['Roll Rate', 'Pitch Rate', 'Yaw Rate']:
    df[col] = df[col].str.split(':').str[-1].str.strip("{}").astype(float)
```
